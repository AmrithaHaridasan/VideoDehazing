```python
#Evaluating the performance of the model on the revide dataset

import itertools 
from pathlib import Path
import os
pred_path = '/content/gdrive/MyDrive/net/pred_FFA_revide'
target_path = '/content/gdrive/MyDrive/net/GT_png'
psn = []
ssi = []
rms = []
for (inp, outp) in zip(os.listdir(pred_path), os.listdir(target_path)):
  #print(inp, outp)
  # print(input_path+inp)
  # p = os.path.join(str(input_path),str(inp))
  # print(p)
  input = cv2.imread(os.path.join(pred_path, inp))
  #print(input)
  output = cv2.imread(os.path.join(target_path, outp))
  ps = psnr(input,output)
  #print(ps)
  psn.append(ps)
  rm = rmse(input,output)
  rms.append(rm)
  ss = ssim(input,output)
  ssi.append(ss)


out_rmse = mean(rms)
out_psnr = mean(psn)
out_ssim = mean(ssi)

print("RMSE ",out_rmse)
print("PSNR ",out_psnr)
print("SSIM ", out_ssim)
```
